*Avec PCA*

```python

#Kmeans avec pca
pca = PCA(n_components=3)
principal_component = pca.fit_transform(x_scaled_data)
start = time.perf_counter()
kmeans_pca = KMeans(n_clusters=3, random_state=42, n_init=10)
# fit the KMeans instance for PCA-transformed data
kmeans_pca.fit(principal_component)
print(f"inertia avec pca: {kmeans_pca.inertia_:.4f}")
end = time.perf_counter()
print(f"temps d'execution avec pca: {end-start:.4f} secondes")
```
